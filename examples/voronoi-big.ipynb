```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib

import numpy as np

from scipy.spatial import Voronoi, voronoi_plot_2d
# import shapely.geometry
# import shapely.ops
```

```python
x_size = 200
y_size = 100
tile_area  = 1
total_area = x_size * y_size
num_tiles  = total_area / tile_area


points = np.random.random( (total_area, 2) )

p2     = np.array([(x * x_size, y * y_size) for (x,y) in points])
vor    = Voronoi(p2)

display(vor.points)
len(points)
```
